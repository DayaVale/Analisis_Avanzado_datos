**5** Realice la segmentaci√≥n para las im√°genes de la carpeta `test_set`. Visualize las segmentaci√≥nes (un s√≥lo grid) y grafique la m√©trica de AIC o BIC (por ejm barras).

```python
from glob import glob
```

```python
images = [Image.open(file) for file in glob('data_test/*.jpg')]
```

```python
plt.figure(figsize=(14,8))
for i in range(len(images)):
    ax=plt.subplot(3,5,i+1) 
    plt.imshow(images[i])  
plt.show()
```

```python
im_total=[]
mult =363*360
for i in images:
    im = np.asarray(i)
    im = np.reshape(im, (mult, 3))
    im_total.append(im)
im_total[0]
```

```python
# Ajuste del modelo con tres componentes 
gmm_model3 = GaussianMixture(n_components=3, covariance_type="tied") 
```

```python
probs=[] 
bics=[]
for i in im_total:
    gmm_model3.fit(i)
    prob3 = gmm_model3.predict_proba(i)
    data3 = pd.DataFrame(prob3) 
    probs.append(data3)
    bics.append(gmm_model3.bic(i))
```

```python
grupitos=[]
for i in probs:
    c_0 = np.reshape(np.array(i[0]), (363,360))
    c_1 = np.reshape(np.array(i[1]), (363,360))
    c_2 = np.reshape(np.array(i[2]), (363,360))
    grupitos.append(c_0) 
    grupitos.append(c_1) 
    grupitos.append(c_2) 
```

```python
plt.figure(figsize=(50,30))
for i in range(len(grupitos)): 
    ax=plt.subplot(5,9,i+1) 
    plt.imshow(grupitos[i],cmap='gray')  
plt.show()
```

```python
eje_x = ['linfocito 99 3071','linfocito 99 3900','linfocito 99 4809','linfocito 99 5059','linfocito 99 6004','linfocito 99 6920','linfocito 99 7617','linfocito 99 9916','linfocito 14 069398','linfocito 14 120881','segmented 979 352','segmented 982 697','segmented 988 734','segmented 991 402','segmented 993 810']
eje_y = bics 
plt.barh(eje_x, eje_y) 
plt.ylabel('imagen') 
plt.xlabel('bics') 
plt.title('Comparacion de bics por imagen') 
plt.show()
```

Las imagenes con mejores bics fueron la de linfocito 14 120881, linfocito 99 7617 y linfocito 99 6004, lo que significa que tuvieron un mejor ajuste y mayor varianza explicada.
