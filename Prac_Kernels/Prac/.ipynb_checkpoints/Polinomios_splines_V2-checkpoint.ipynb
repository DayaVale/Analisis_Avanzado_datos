**Definición de la función de utilidad.**

```python
def ortho_poly_fit(x, degree = 1):
    '''
    Convert data into orthogonal basis for polynomial regression by QR decomposition.
    '''
    n = degree + 1
    x = np.asarray(x).flatten()
    if(degree >= len(np.unique(x))):
            stop("'degree' must be less than number of unique points")
    xbar = np.mean(x)
    x = x - xbar
    X = np.fliplr(np.vander(x, n))
    q,r = np.linalg.qr(X)

    z = np.diag(np.diag(r))
    raw = np.dot(q, z)

    norm2 = np.sum(raw**2, axis=0)
    alpha = (np.sum((raw**2)*np.reshape(x,(-1,1)), axis=0)/norm2 + xbar)[:degree]
    Z = raw / np.sqrt(norm2)
    return Z, norm2, alpha
```

### Regresión polinomial y funciones escalonadas

**Regresión polinomial de grado 4 sobre X ortogonalizada.**

```python
# Polynomial regression of degree 4 on orthogonalized X.
X4_ortho = ortho_poly_fit(wage[['age']], degree=4)[0]
X4_ortho[:,0]=1  # Replace constant column with 1s for Intercept estimation.
poly4_ortho = sm.GLS(wage['wage'], X4_ortho).fit()
print(poly4_ortho.summary())
```

**Regresión polinomial de grado 4 en X en bruto sin ortogonalización.**

```python
# Polynomial regression of degree 4 on raw X without orthogonalization.
X4 = PolynomialFeatures(degree=4).fit_transform(wage[['age']])
poly4 = sm.GLS(wage['wage'], X4).fit()
print(poly4.summary())
```

**Predección sobre una grid de edad.**

```python
# Generate a sequence of age values spanning the range
age_grid = np.arange(wage.age.min(), wage.age.max()).reshape(-1,1)

# Generate test data
X_test = PolynomialFeatures(4).fit_transform(age_grid)

# Predict the value of the generated ages
y_pred = poly4.predict(X_test)

# Plot
fig, ax = plt.subplots(figsize=(8,6))
fig.suptitle('Degree-4 Polynomial Regression', fontsize=14)

# Scatter plot with polynomial regression line
plt.scatter(wage.age, wage.wage, facecolor='None', edgecolor='k', alpha=0.5)
plt.plot(age_grid, y_pred, color = 'b')
ax.set_ylim(ymin=0)
plt.xlabel('Age')
plt.ylabel('Wage')
plt.show()
```

```python
# Polynomial regression of degree 5 on orthogonalized X.
X5_ortho = ortho_poly_fit(wage[['age']], degree=5)[0]
X5_ortho[:,0]=1  # Replace constant column with 1s for Intercept estimation.
poly5_ortho = sm.GLS(wage['wage'], X5_ortho).fit()
print(poly5_ortho.summary())
```

### Design matrix (dmatrix) y `patsy`

En estadística, una matriz de diseño, también conocida como matriz modelo o matriz regresiva y a menudo denotada por X, es una matriz de valores de variables explicativas de un conjunto de objetos. Cada fila representa un objeto individual, con las columnas sucesivas correspondientes a las variables y sus valores específicos para ese objeto.

La matriz de diseño contiene datos sobre las variables independientes (también llamadas variables explicativas) en modelos estadísticos que intentan explicar los datos observados sobre una variable de respuesta (a menudo llamada variable dependiente) en términos de las variables explicativas.

La matriz de diseño se define como una matriz $X$ tal que $X_{ij}$ (la j-ésima columna de la i-ésima fila de $X$) representa el valor de la j-ésima variable asociada con el i-ésimo objeto.

Por tanto, un modelo de regresión que es una combinación lineal de las variables explicativas puede representarse mediante la multiplicación de matrices como

$$y=X\beta$$
donde X es la matriz de diseño, $\beta$ es un vector de los coeficientes del modelo (uno para cada variable) e y es el vector de resultados predichos para cada objeto.

Regresión múltiple
Esta sección contiene un ejemplo de regresión múltiple con dos covariables (variables explicativas): w y x. Nuevamente, suponga que los datos constan de siete observaciones, y que para cada valor observado que se va a predecir ({\ displaystyle y _{i}} y_ {i}), los valores wi y xi de las dos covariables también son observado. El modelo a considerar es

${\displaystyle y_{i}=\beta _{0}+\beta _{1}w_{i}+\beta _{2}x_{i}+\varepsilon _{i}}{\displaystyle y_{i}=\beta _{0}+\beta _{1}w_{i}+\beta _{2}x_{i}+\varepsilon _{i}}$
Este modelo se puede escribir en términos matriciales como

${\displaystyle {\begin{bmatrix}y_{1}\\y_{2}\\y_{3}\\y_{4}\\y_{5}\\y_{6}\\y_{7}\end{bmatrix}}={\begin{bmatrix}1&w_{1}&x_{1}\\1&w_{2}&x_{2}\\1&w_{3}&x_{3}\\1&w_{4}&x_{4}\\1&w_{5}&x_{5}\\1&w_{6}&x_{6}\\1&w_{7}&x_{7}\end{bmatrix}}{\begin{bmatrix}\beta _{0}\\\beta _{1}\\\beta _{2}\end{bmatrix}}+{\begin{bmatrix}\varepsilon _{1}\\\varepsilon _{2}\\\varepsilon _{3}\\\varepsilon _{4}\\\varepsilon _{5}\\\varepsilon _{6}\\\varepsilon _{7}\end{bmatrix}}}{\displaystyle {\begin{bmatrix}y_{1}\\y_{2}\\y_{3}\\y_{4}\\y_{5}\\y_{6}\\y_{7}\end{bmatrix}}={\begin{bmatrix}1&w_{1}&x_{1}\\1&w_{2}&x_{2}\\1&w_{3}&x_{3}\\1&w_{4}&x_{4}\\1&w_{5}&x_{5}\\1&w_{6}&x_{6}\\1&w_{7}&x_{7}\end{bmatrix}}{\begin{bmatrix}\beta _{0}\\\beta _{1}\\\beta _{2}\end{bmatrix}}+{\begin{bmatrix}\varepsilon _{1}\\\varepsilon _{2}\\\varepsilon _{3}\\\varepsilon _{4}\\\varepsilon _{5}\\\varepsilon _{6}\\\varepsilon _{7}\end{bmatrix}}}$

Aquí, la matriz de 7 × 3 del lado derecho es la matriz de diseño.

### Splines

**Cubic splines**

```python
# Specifying 3 knots
transformed_3knots = dmatrix("bs(wage.age, knots=(25,40,60), degree=3, include_intercept=False)",
                         {"wage.age": wage.age}, return_type='dataframe')

# Build a regular linear model from the splines
spln_3knots = sm.GLM(wage.wage, transformed_3knots).fit()
pred_3knots = spln_3knots.predict(dmatrix("bs(age_grid, knots=(25,40,60), degree=3, include_intercept=False)",
                                          {"age_grid": age_grid}, return_type='dataframe'))
spln_3knots.params
```

```python
# Specifying 6 degrees of freedom 
transformed_deg6 = dmatrix("bs(wage.age, df=6, include_intercept=False)",
                        {"wage.age": wage.age}, return_type='dataframe')
spln_deg6 = sm.GLM(wage.wage, transformed_deg6).fit()
pred_deg6 = spln_deg6.predict(dmatrix("bs(age_grid, df=6, degree=3, include_intercept=False)",
                                  {"age_grid": age_grid}, return_type='dataframe'))
spln_deg6.params
```

**Natural splines**

```python
# Specifying 4 degrees of freedom
transformed_deg4 = dmatrix("cr(wage.age, df=4)", {"wage.age": wage.age}, return_type='dataframe')
spln_deg4 = sm.GLM(wage.wage, transformed_deg4).fit()
pred_deg4 = spln_deg4.predict(dmatrix("cr(age_grid, df=4)", {"age_grid": age_grid}, return_type='dataframe'))
spln_deg4.params
```

```python
# Plot splines
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(wage.age, wage.wage, facecolor='None', edgecolor='k', alpha=0.3)
plt.plot(age_grid, pred_3knots, color='b', label='Specifying three knots')
plt.plot(age_grid, pred_deg6, color='r', label='Specifying df=6')
plt.plot(age_grid, pred_deg4, color='g', label='Natural spline df=4')
[plt.vlines(i , 0, 350, linestyles='dashed', lw=2, colors='b') for i in [25,40,60]]
plt.legend(bbox_to_anchor=(1.5, 1.0))
plt.xlim(15,85)
plt.ylim(0,350)
plt.xlabel('age')
plt.ylabel('wage');
```

### Regresión con splines (no suaves) con múltiples variables

####  Para probar la regresión, dividimos el wage dataset en training y test sets

```python
wage2 = pd.get_dummies(wage,drop_first=True)
wage2.columns
```

```python
wage_train, wage_test = train_test_split(wage2, test_size=0.25, random_state=123)
print(wage_train.shape)
print(wage_test.shape)
wage_train.head()
```

A continuación realizamos un ajuste con spline natural cúbico usando los datos de entrenamiento y realizando la predicción sobre el test set. La métrica usada es el *RMSE* (root mean squared error).

```python
# Specifying 4 degrees of freedom
transformed_deg4 = dmatrix("cr(wage_train.age, df=4)", {"wage_train.age": wage_train.age}, return_type='dataframe')
spln_deg4 = sm.GLM(wage_train.wage, transformed_deg4).fit()
pred_deg4 = spln_deg4.predict(dmatrix("cr(wage_test.age, df=4)", {"wage_test.age": wage_test.age}, return_type='dataframe'))

print("El RMSE es:",np.sqrt(mean_squared_error(wage_test.wage, pred_deg4)) )
```

```python
wage['wage'].describe()
```

#### El siguiente es un ejemplo para implicar múltiples variables en la regresión con un spline natural cúbico (`cr()` en patsy). Para hacerlo, se ha usado la multiplicación tensorial de diferentes bases de funciones, que usando `patsy` se realiza mediante `te()`. 
**Una de las variables es categórica y ha sido codificada anteriormente con `get_dummies` para poder usarla dentro del spline.**

La documentación completa sobre splines usando `patsy` para definir la transformación, la pueden encontrar [aquí](https://patsy.readthedocs.io/en/latest/spline-regression.html).


```python
deg=4 # Grados de libertad

#Definicion de la transformacion
transformacion = dmatrix(
    f"te(cr(wage_train.age, df={deg}), cr(wage_train.year, df={deg}),cr(jobclass,df={deg}))", # se definen los splines
                           {"wage_train.age": wage_train.age, "wage_train.year": wage_train.year,
                            "jobclass":wage_train['jobclass_2. Information']} # Se definen los datos
)

model_spline = sm.GLM(wage_train.wage, transformacion).fit() # Train

# Prediccion
prediccion = model_spline.predict(
    dmatrix(f"te(cr(wage_test.age, df={deg}), cr(wage_test.year, df={deg}), cr(jobclass,df={deg}) )", # Splines
            {"wage_test.age": wage_test.age, "wage_test.year": wage_test.year,
             "jobclass":wage_test['jobclass_2. Information']}, # datos
            return_type='dataframe') )

print("El RMSE es:",np.sqrt(mean_squared_error(wage_test.wage, prediccion)) )
```

### Splines suavizados

Para implementar los splines suavizados, usaremos la librería `csaps`. Si no está instalada, deben ejecutar la instrucción `pip install -U csaps` en el terminal o si están en colab (o en el jupyter notebook) en una celda: `!pip install -U csaps`.

```python
from csaps import csaps
```

El paquete `csaps` define la penalización de regularización de manera ligeramente diferente a la vista en clase: acepta valores entre 0 y 1, donde 0 es una línea recta y 1 es una interpolación spline cúbica natural. El paquete también requiere que preparemos los datos antes de ajustar el modelo. Solo puede manejar un valor de 𝑌 por cada 𝑋. Por lo tanto, agruparemos por `age` y agregaremos `wage` utilizando la mediana.

Primero debemos ordenar de forma ascendente la variable independiente y la dependiente de acuerdo a ésta.

```python
wage2 = wage.sort_values("age").groupby("age")["wage"].median().reset_index()
wage2.head()
```

#### Spline suavizado en 1D

```python
sp = csaps(xdata=wage2.age, ydata=wage2.wage, smooth=0.1) # Ajusta el modelo de smooth spline

pred_smooth_sp = sp(age_grid) # prediccion sobre nuevos valores de edad (un grid)
```

```python
# Grafica el spline suavizado
plt.plot(age_grid, pred_smooth_sp, color='r', label='Smoothing spline')
plt.legend()
plt.scatter(wage['age'], wage['wage'], facecolor='None', edgecolor='k', alpha=0.05)
plt.xlabel('Age')
plt.ylabel('Wage')
plt.show()
```

### Spline suavizado multidimensional

El siguiente es un ejemplo de un spline usando `csaps` para dos dimensiones, pero con base en un grid de valores

```python
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(123)
xdata = [np.linspace(-3, 3, 41), np.linspace(-3.5, 3.5, 31)]
i, j = np.meshgrid(*xdata, indexing='ij')
ydata = (3 * (1 - j)**2. * np.exp(-(j**2) - (i + 1)**2)
         - 10 * (j / 5 - j**3 - i**5) * np.exp(-j**2 - i**2)
         - 1 / 3 * np.exp(-(j + 1)**2 - i**2))
ydata = ydata + (np.random.randn(*ydata.shape) * 0.75)

ydata_s = csaps(xdata, ydata, xdata, smooth=0.988)

fig = plt.figure(figsize=(14, 9))
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('none')
c = [s['color'] for s in plt.rcParams['axes.prop_cycle']]
ax.plot_wireframe(j, i, ydata, linewidths=0.5, color=c[0], alpha=0.5)
ax.scatter(j, i, ydata, s=10, c=c[0], alpha=0.5)
ax.plot_surface(j, i, ydata_s, color=c[1], linewidth=0, alpha=1.0)
ax.view_init(elev=9., azim=290)

plt.show()
```
