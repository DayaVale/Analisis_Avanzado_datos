# Trucos de kernel y reducción de dimensionalidad no lineal a través de PCA

**La mayoría de los algoritmos de aprendizaje automático se han desarrollado y validado estadísticamente para datos que son separables linealmente. Algunos ejemplos populares son los clasificadores lineales como Support Vector Machines (SVM) o el Análisis de componentes principales (PCA) (estándar) para la reducción de dimensionalidad. Sin embargo, la mayoría de los datos del mundo real requieren métodos no lineales para realizar tareas que implican el análisis y el descubrimiento de patrones con éxito.**

**El enfoque de este notebook es presentar brevemente la idea de los métodos del kernel e implementar un kernel Gaussiano de base radial (RBF) que se utiliza para realizar una reducción de dimensionalidad no lineal a través del análisis de componentes principales del kernel RBF (kPCA).**

## Principal Component Analysis

El propósito principal del análisis de componentes principales (PCA) es el análisis de datos para identificar patrones que representen "bien" los datos. Los componentes principales pueden entenderse como nuevos ejes del conjunto de datos que maximizan la varianza a lo largo de esos ejes (los vectores propios de la matriz de covarianza). En otras palabras, PCA tiene como objetivo encontrar los ejes con varianzas máximas a lo largo de los cuales los datos están más dispersos.

![pca overview](pca_1.png)

## PCA and linear dimensionality reduction

Una aplicación común de PCA es reducir las dimensiones del conjunto de datos con una pérdida mínima de información.
Aquí, el conjunto de datos completo (*d* dimensiones) se proyecta en un nuevo subespacio (*k* dimensiones donde *k* <*d*).
Este método de proyección es útil para reducir los costos computacionales y el error de estimación de parámetros ("maldición de la dimensionalidad").

El enfoque de PCA estándar se puede resumir en cinco pasos simples:

1. Estandarizar los datos.
2. Obtener los autovectores y autovalores de la matriz de covarianza o matriz de correlación, o realizar la descomposición de vectores singulares.
3. Ordenar los autovalores en orden descendente y escoger los $k$-ésimos autovectores que correspondan a los $k$-ésimos autovalores más grandes, donde $k$ es el número de dimensiones del subespacio de la nueva característica ($k \le d$) .
4. Construir la matriz de proyección $\mathbf{W}$ a partir de los vectores propios de $k$ seleccionados.
5. Transformar el conjunto de datos original $\mathbf{X}$ a través de $\mathbf{W}$ para obtener un subespacio de características $k$-dimensional $\mathbf{Y}$.

## Reducción de dimensionalidad no lineal

El enfoque de PCA "clásico" descrito anteriormente es una técnica de proyección lineal que funciona bien si los datos son linealmente separables. Sin embargo, en el caso de datos linealmente inseparables, se requiere una técnica no lineal si la tarea es reducir la dimensionalidad de un conjunto de datos.

## Funciones de kernel y el truco del kernel

![](kernelTrick_transform.png)

La idea básica para tratar con datos linealmente inseparables es proyectarlos en un espacio dimensional superior donde se vuelven linealmente separables. Llamemos a esta función de mapeo no lineal $\phi$ para que el mapeo de un ejemplo de $\mathbf{x}$ se pueda escribir como $\mathbf{x} \rightarrow \phi (\mathbf{x})$, que se llama "función del núcleo".

Para transformar los ejemplos $x \in \mathbb{R}^{d}$ en este subespacio de mayor dimensión $k$, definimos una función de mapeo no lineal, $\phi$:
$$
\phi: \quad \mathbb{R}^{d} \rightarrow \mathbb{R}^{k} \quad(k \gg d)
$$
Podemos pensar en $\phi$ como una función que crea combinaciones no lineales de las características originales para mapear el conjunto de datos original $d$-dimensional a un espacio de características $k$-dimensional más grande.

Por ejemplo, si tuviéramos un vector de características $x \in \mathbb{R}^{d}(\boldsymbol{x}$ es un vector columna que consta de $d$ características) con dos dimensiones $(d=2)$, un mapeo potencial a un espacio 3D podría ser:
$$
\begin{array}{c}
\boldsymbol{x}=\left[x_{1}, x_{2}\right]^{T} \\
\downarrow \phi \\
\mathbf{z}=\left[x_{1}^{2}, \sqrt{2 x_{1} x_{2}}, x_{2}^{2}\right]^{T}
\end{array}
$$

En otras palabras, realizamos un mapeo no lineal a través de KPCA que transforma los datos a un espacio de mayor dimensión. Luego usamos PCA estándar en este espacio de mayor dimensión para proyectar los datos de nuevo en un espacio de menor dimensión donde los ejemplos pueden separarse mediante un clasificador lineal (bajo la condición de que los ejemplos se puedan separar por densidad en el espacio de entrada). Sin embargo, una desventaja de este enfoque es que es computacionalmente muy costoso, y aquí es donde usamos **el truco del kernel**. Usando el truco del kernel, podemos calcular la similitud entre dos vectores de características de alta dimensión en el espacio de características original.

Antes de continuar con más detalles sobre el truco del kernel para abordar este problema computacionalmente costoso, pensemos en el enfoque estándar de PCA. Calculamos la covarianza entre dos características, $k$ y $j$, de la siguiente manera:
$$
\sigma_{j k}=\frac{1}{n} \sum_{i=1}^{n}\left(x_{j}^{(i)}-\mu_{j}\right)\left(x_{k}^{(i)}-\mu_{k}\right)
$$
Dado que la estandarización de características las centra en la media cero, por ejemplo, $\mu_{j}=0$ y $\mu_{k}=0$, podemos simplificar esta ecuación de la siguiente manera:
$$
\sigma_{j k}=\frac{1}{n} \sum_{i=1}^{n} x_{j}^{(i)} x_{k}^{(i)}
$$

Tenga en cuenta que la ecuación anterior se refiere a la covarianza entre dos características; ahora, escribamos la ecuación general para calcular la matriz de covarianza, $\Sigma$:
$$
\Sigma=\frac{1}{n} \sum_{i=1}^{n} \boldsymbol{x}^{(i)} \boldsymbol{x}^{(i)^{T}}
$$
Bernhard Scholkopf (1997) generalizó este enfoque para que podamos reemplazar los productos escalares entre muestras en el espacio de características original con las combinaciones de características no lineales a través de $\phi$:
$$
\Sigma=\frac{1}{n} \sum_{i=1}^{n} \phi\left(\boldsymbol{x}^{(i)}\right) \phi\left(\boldsymbol{x}^{(i)}\right)^{T}
$$

Para obtener los vectores propios, es decir las componentes principales, de esta matriz de covarianza, tenemos que resolver la siguiente ecuación:
$$\Sigma \boldsymbol{v}=\lambda \boldsymbol{v} $$
$$\Rightarrow \frac{1}{n} \sum_{i=1}^{n} \phi\left(\boldsymbol{x}^{(i)}\right) \phi\left(\boldsymbol{x}^{(i)}\right)^{T} \boldsymbol{v}=\lambda \boldsymbol{v}$$
$$\Rightarrow v=\frac{1}{n \lambda} \sum_{i=1}^{n} \phi\left(\boldsymbol{x}^{(i)}\right) \phi\left(\boldsymbol{x}^{(i)}\right)^{T} \boldsymbol{v}=\frac{1}{n} \sum_{i=1}^{n} \boldsymbol{a}^{(i)} \phi\left(\boldsymbol{x}^{(i)}\right)$$

Aquí, $\lambda$ y $\boldsymbol{v}$ son los autovalores y los autovectores de la matriz de covarianza, $\Sigma$ y $\boldsymbol{a}$ se pueden obtener extrayendo los autovectores de la matriz del kernel (similitud), $\boldsymbol{K}$, como se demuestra a continuación.

#### Derivando la matriz del kernel
La derivación de la matriz del kernel se puede mostrar de la siguiente manera. Primero, escribamos la matriz de covarianza en notación matricial, donde $\phi(X)$ es una matriz de $n \times k$ dimensiones:
$$
\Sigma=\frac{1}{n} \sum_{i=1}^{n} \phi\left(\boldsymbol{x}^{(i)}\right) \phi\left(\boldsymbol{x}^{(i)}\right)^{T}=\frac{1}{n} \phi(\boldsymbol{X})^{T} \phi(\boldsymbol{X})
$$
Ahora, podemos escribir la ecuación de vectores propios de la siguiente manera:
$$
v=\frac{1}{n} \sum_{i=1}^{n} a^{(i)} \phi\left(\boldsymbol{x}^{(i)}\right)=\frac{1}{n} \phi(\boldsymbol{X})^{T} \boldsymbol{a}
$$
Reemplazando las ecuaciones anteriores en $\Sigma \boldsymbol{v}=\lambda \boldsymbol{v}$, obtenemos:
$$
\frac{1}{n} \phi(\boldsymbol{X})^{T} \phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T} \boldsymbol{a}=\lambda \phi(\boldsymbol{X})^{T} \boldsymbol{a}
$$
Multiplicando por $\phi(\boldsymbol{X})$ en ambos lados, produce el siguiente resultado:
$$
\begin{array}{l}
\frac{1}{n} \phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T} \phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T} \boldsymbol{a}=\lambda \phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T} \boldsymbol{a} \\
\Rightarrow \frac{1}{n} \phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T} \boldsymbol{a}=\lambda \boldsymbol{a} \\
\Rightarrow \frac{1}{n} \boldsymbol{K} \boldsymbol{a}=\lambda \boldsymbol{a}
\end{array}
$$

Aquí, $\boldsymbol{K}$ es la matriz de similitud (kernel):
$$
\boldsymbol{K}=\phi(\boldsymbol{X}) \phi(\boldsymbol{X})^{T}
$$

Así, podemos usar el truco del kernel para evitar calcular los productos de puntos por pares de las muestras, $\boldsymbol{x}$, bajo $\phi$ explícitamente usando una función del kernel, $\kappa$, de modo que no necesitemos calcular los autovectores explícitamente:
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\phi\left(\boldsymbol{x}^{(i)}\right)^{T} \phi\left(\boldsymbol{x}^{(j)}\right)
$$

En otras palabras, lo que obtenemos después de KPCA son las muestras ya proyectadas en las componentes respectivas, en lugar de construir una matriz de transformación como en el enfoque estándar de PCA. Básicamente, la función kernel (o simplemente kernel) puede entenderse como una función que calcula un producto escalar entre dos vectores, es decir, es una medida de similitud.

Los kernels más utilizados son los siguientes:
- El kernel polinomial:
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\left(\boldsymbol{x}^{(i)^{T}} \boldsymbol{x}^{(j)}+\theta\right)^{p}
$$
Aquí, $\theta$ es el umbral y $p$ es la potencia que debe especificar el usuario.
- El kernel tangente hiperbólico (sigmoide):
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\tanh \left(\eta \boldsymbol{x}^{(i)^{T}} \boldsymbol{x}^{(j)}+\theta\right)
$$
- La función de base radial (RBF) o kernel gaussiano, que usaremos en los ejemplos siguientes:
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\exp \left(-\frac{\left\|\boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}\right\|^{2}}{2 \sigma^{2}}\right)
$$
Este kernel RBF se escribe a menudo de la siguiente forma, introduciendo la variable $\gamma=\frac{1}{2 \sigma^{2}}$.
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\exp \left(-\gamma\left\|\boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}\right\|^{2}\right)
$$

To summarize what we have learned so far, we can define the following three steps to implement an RBF KPCA:

1. We compute the kernel (similarity) matrix, $\boldsymbol{K}$, where we need to calculate the following:
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\exp \left(-\gamma\left\|\boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}\right\|^{2}\right)
$$ We do this for each pair of examples:
$$
\boldsymbol{K}=\left[\begin{array}{cccc}
\kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(1)}\right) & \kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(n)}\right) \\
\kappa\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(1)}\right) & \left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(n)}\right) \\
\vdots & \vdots & \ddots & \vdots \\
\kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(1)}\right) & \kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(n)}\right)
\end{array}\right]
$$
For example, if our dataset contains 100 training examples, the symmetric kernel matrix of the pairwise similarities would be $100 \times 100$-dimensional.

2. We center the kernel matrix, $\boldsymbol{K}$, using the following equation:
$$
K^{\prime}=K-1_{n} K-K 1_{n}+1_{n} K 1_{n}
$$
Here, $\mathbf{1}_{n}$ is an $n \times n$-dimensional matrix (the same dimensions as the kernel matrix) where all values are equal to $\frac{1}{n}$.

3. We collect the top $k$ eigenvectors of the centered kernel matrix based on their corresponding eigenvalues, which are ranked by decreasing magnitude. In contrast to standard PCA, the eigenvectors are not the principal component axes, but the examples already projected onto these axes.

At this point, you may be wondering why we need to center the kernel matrix in the second step. We previously assumed that we are working with standardized data, where all features have mean zero when we formulate the covariance matrix and replace the dot-products with the nonlinear feature combinations via $\phi .$ Thus, the centering of the kernel matrix in the second step becomes necessary, since we do not compute the new feature space explicitly so that we cannot guarantee that the new feature space is also centered at zero.
In the next section, we will put those three steps into action by implementing a KPCA in Python.

## Implementando el kernel RBF PCA paso a paso

Para resumir lo que hemos aprendido hasta ahora, podemos definir los siguientes tres pasos para implementar un RBF KPCA:

1. Calculamos la matriz del kernel (similitud), $\boldsymbol{K}$, donde necesitamos calcular lo siguiente:
$$
\kappa\left(\boldsymbol{x}^{(i)}, \boldsymbol{x}^{(j)}\right)=\exp \left(-\gamma\left\|\boldsymbol{x}^{(i)}-\boldsymbol{x}^{(j)}\right\|^{2}\right)
$$ Hacemos esto para cada par de ejemplos:
$$
\boldsymbol{K}=\left[\begin{array}{cccc}
\kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(1)}\right) & \kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(1)}, \boldsymbol{x}^{(n)}\right) \\
\kappa\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(1)}\right) & \left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(2)}, \boldsymbol{x}^{(n)}\right) \\
\vdots & \vdots & \ddots & \vdots \\
\kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(1)}\right) & \kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(2)}\right) & \cdots & \kappa\left(\boldsymbol{x}^{(n)}, \boldsymbol{x}^{(n)}\right)
\end{array}\right]
$$
Por ejemplo, si nuestro dataset contiene 100 ejemplos de entrenamiento, la matriz de kernel simétrica de las similitudes por pares sería de dimensión $100 \times 100$.

2. Centramos la matriz del kernel, $\boldsymbol{K}$, usando la siguiente ecuación:
$$
\boldsymbol{K}^{\prime}=\boldsymbol{K}-\mathbf{1}_{n} \boldsymbol{K}-\boldsymbol{K} \mathbf{1}_{n}+\mathbf{1}_{n} \boldsymbol{K} \mathbf{1}_{n}
$$
Aquí, $\mathbf{1}_{n}$ es una matriz de dimensión $n \times n$ (las mismas dimensiones que la matriz del kernel) donde todos los valores son iguales a $\frac{1}{n}$.

3. Recopilamos los mejores $k$ autovectores de la matriz del kernel centrada en función de sus autovalores correspondientes, que se organizan de forma decreciente en magnitud. **A diferencia del PCA estándar, los autovectores no son los ejes o componentes principales, sino las muestras ya proyectadas sobre estos ejes.**

En este punto, es posible que se pregunte por qué necesitamos centrar la matriz del kernel en el segundo paso. Anteriormente asumimos que estamos trabajando con datos estandarizados, donde todas las características tienen una media de cero cuando formulamos la matriz de covarianza y reemplazamos los productos punto con las combinaciones de características no lineales a través de $\phi .$ Por lo tanto, el centrado de la matriz del kernel en el segundo paso se vuelve necesario, ya que no calculamos el nuevo espacio de características explícitamente, por lo que no podemos garantizar que el nuevo espacio de características también esté centrado en cero.
A continuación, pondremos esos tres pasos en acción mediante la implementación de un KPCA en Python.

```python
from scipy.spatial.distance import pdist, squareform
from scipy.linalg import eigh
import numpy as np
from numpy import exp

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.

    Parameters
    ------------
    X: {NumPy ndarray}, shape = [n_examples, n_features]
        
    gamma: float
      Tuning parameter of the RBF kernel
        
    n_components: int
      Number of principal components to return

    Returns
    ------------
     X_pc: {NumPy ndarray}, shape = [n_examples, k_features]
       Projected dataset   

    """
    # Calculate pairwise squared Euclidean distances
    # in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')

    # Convert pairwise distances into a square matrix.
    mat_sq_dists = squareform(sq_dists)

    # Compute the symmetric kernel matrix.
    K = exp(-gamma * mat_sq_dists)

    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Obtaining eigenpairs from the centered kernel matrix
    # scipy.linalg.eigh returns them in ascending order
    eigvals, eigvecs = eigh(K)
    eigvals, eigvecs = eigvals[::-1], eigvecs[:, ::-1]

    # Collect the top k eigenvectors (projected examples)
    X_pc = np.column_stack([eigvecs[:, i]
                            for i in range(n_components)])

    return X_pc
```

# Ejemplos de RBF Kernel PCA

En esta sección, aplicaremos PCA con el  kernel RBF a diferentes datos de muestra no lineales para realizar la reducción de dimensionalidad.

## Ejemplo: formas de media luna

Comenzaremos con un ejemplo simple de 2 formas de media luna generadas por [make_moons](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_moons.html) de [scikit-learn](http://scikit-learn.org/stable/index.html).

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, random_state=123)

plt.figure(figsize=(8,6))

plt.scatter(X[y==0, 0], X[y==0, 1], color='red', alpha=0.5)
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue', alpha=0.5)

plt.title('A nonlinear 2Ddataset')
plt.ylabel('y coordinate')
plt.xlabel('x coordinate')
plt.show()
```

### PCA Lineal

Dado que las dos formas de media luna son linealmente inseparables, esperamos que el PCA "clásico" no nos proporcione una "buena" representación de los datos en el espacio 1D. Aquí, usaremos la clase [`PCA`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) que se implementa en scikit-learn para realizar la reducción de dimensionalidad.

```python
from sklearn.decomposition import PCA

scikit_pca = PCA(n_components=2)
X_spca = scikit_pca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_spca[y==0, 0], X_spca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_spca[y==1, 0], X_spca[y==1, 1], color='blue', alpha=0.5)

plt.title('First 2 principal components after Linear PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

Y ahora para una sólo componente principal:

```python
plt.figure(figsize=(8,6))
plt.scatter(X_spca[y==0, 0], np.zeros((50,1)), color='red', alpha=0.5)
plt.scatter(X_spca[y==1, 0], np.zeros((50,1)), color='blue', alpha=0.5)

plt.title('First principal component after Linear PCA')
plt.xlabel('PC1')

plt.show()
```

Como podemos ver, los componentes principales resultantes no producen un subespacio donde los datos estén bien separados linealmente. Tenga en cuenta que PCA es un método no supervisado y no "considera" las etiquetas de clase para maximizar la varianza en contraste con el Análisis discriminante lineal. Aquí, los colores azul y rojo solo se agregan con fines de visualización para indicar el grado de separación.

### PCA del kernel RBF gaussiano

A continuación, realizaremos la reducción de dimensionalidad a través de PCA del kernel RBF en nuestros datos de media luna. La elección de $\gamma$ depende del conjunto de datos y se puede obtener mediante técnicas de ajuste de hiperparámetros como Grid Search. El ajuste de hiperparámetros es un tema amplio en sí mismo, y aquí solo usaré un valor $\gamma$ que encontré para producir "buenos" resultados.

```python
X_pc = rbf_kernel_pca(X, gamma=15, n_components=2)

plt.figure(figsize=(8,6))
plt.scatter(X_pc[y==0, 0], X_pc[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_pc[y==1, 0], X_pc[y==1, 1], color='blue', alpha=0.5)

plt.title('First 2 principal components after RBF Kernel PCA')
plt.text(-0.18, 0.18, 'gamma = 15', fontsize=12)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pc[y==0, 0], np.zeros((50)), color='red', alpha=0.5)
plt.scatter(X_pc[y==1, 0], np.zeros((50)), color='blue', alpha=0.5)

plt.title('First principal component after RBF Kernel PCA')
plt.text(-0.14, 0.04, 'gamma = 15', fontsize=12)
plt.xlabel('PC1')
plt.show()
```

Podemos ver claramente que la proyección a través de RBF kernel PCA produjo un subespacio donde las clases están bien separadas. Dicho subespacio se puede utilizar como entrada para modelos de clasificación lineal, como Support Vector Machines o clasificadores de naive Bayes, por ejemplo.

### scikit RBF kernel PCA

Para nuestra comodidad, ya existe una implementación de [`KernelPCA`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html) en scikit-learn.
Confirmemos que los resultados de nuestra implementación son consistentes con el enfoque de scikit-learn.

```python
from sklearn.decomposition import KernelPCA

scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_skernpca[y==0, 0], X_skernpca[y==0, 1], color='red', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], X_skernpca[y==1, 1], color='blue', alpha=0.5)

plt.text(-0.38, 0.37, 'gamma = 15', fontsize=12)
plt.title('First 2 principal components after RBF Kernel PCA via scikit-learn')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
scikit_kpca = KernelPCA(n_components=1, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)

plt.figure(figsize=(8,6))
plt.scatter(X_skernpca[y==0, 0], np.zeros((50,1)), color='red', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], np.zeros((50,1)), color='blue', alpha=0.5)
plt.text(-0.38, 0.04, 'gamma = 15', fontsize=12)
plt.title('First principal component after RBF Kernel PCA')
plt.xlabel('PC1')
plt.show()
```

# Proyectando nuevos datos

Hasta ahora, hemos estado proyectando un conjunto de datos en un nuevo subespacio de características. Sin embargo, en una aplicación real, generalmente estamos interesados en mapear nuevos puntos de datos al mismo subespacio de nuevas características (por ejemplo, si estamos trabajando con un entrenamiento y un conjunto de datos de prueba en tareas de clasificación).

Recuerde, cuando calculamos los vectores propios \\( \mathbf{a} \\) de la matriz del kernel centrada, esos valores ya eran en realidad los puntos de datos proyectados en el eje del componente principal \\( \boldsymbol{v} \\).

Si queremos proyectar un nuevo punto de datos \\( \boldsymbol{x}^{\prime} \\) en este eje del componente principal, necesitaríamos calcular $\phi\left(\boldsymbol{x}^{\prime}\right)^{T} \boldsymbol{v}$.

Afortunadamente, no tenemos que calcular $\phi\left(\boldsymbol{x}^{\prime}\right)^{T} \boldsymbol{v}$ explícitamente, sino que usamos el truco del kernel para calcular el kernel RBF entre el nuevo punto de datos \\( \boldsymbol{x}^{\prime} \\) y cada $i$-ésimo punto de datos en el conjunto de datos de entrenamiento:

$$
\begin{aligned}
\phi\left(\boldsymbol{x}^{\prime}\right)^{T} \boldsymbol{v} &=\sum_{i} a^{(i)} \phi\left(\boldsymbol{x}^{\prime}\right)^{T} \phi\left(\boldsymbol{x}^{(i)}\right) \\
&=\sum_{i} a^{(i)} \kappa\left(\boldsymbol{x}^{\prime}, \boldsymbol{x}^{(i)}\right)
\end{aligned}
$$


Aquí, los vectores propios, $\boldsymbol{a}$ y los valores propios, $\lambda$, de la matriz del núcleo, $\boldsymbol{K}$, satisfacen la siguiente condición en la ecuación:
$$
\boldsymbol{K} \boldsymbol{a}=\lambda \boldsymbol{a}
$$
Después de calcular la similitud entre las nuevas muestras y las muestras del conjunto de datos de entrenamiento, tenemos que normalizar el autovector, $\boldsymbol{a}$, por su autovalor. Por lo tanto, modifiquemos la función *rbf _kernel_ pca* que implementamos anteriormente para que también devuelva los valores propios de la matriz del kernel:

```python
from scipy.spatial.distance import pdist, squareform
from numpy import exp
from scipy.linalg import eigh
import numpy as np

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.

    Parameters
    ------------
    X: {NumPy ndarray}, shape = [n_examples, n_features]
        
    gamma: float
      Tuning parameter of the RBF kernel
        
    n_components: int
      Number of principal components to return

    Returns
    ------------
     alphas: {NumPy ndarray}, shape = [n_examples, k_features]
       Projected dataset 
     
     lambdas: list
       Eigenvalues

    """
    # Calculate pairwise squared Euclidean distances
    # in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')

    # Convert pairwise distances into a square matrix.
    mat_sq_dists = squareform(sq_dists)

    # Compute the symmetric kernel matrix.
    K = exp(-gamma * mat_sq_dists)

    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Obtaining eigenpairs from the centered kernel matrix
    # scipy.linalg.eigh returns them in ascending order
    eigvals, eigvecs = eigh(K)
    eigvals, eigvecs = eigvals[::-1], eigvecs[:, ::-1]

    # Collect the top k eigenvectors (projected examples)
    alphas = np.column_stack([eigvecs[:, i]
                              for i in range(n_components)])

    # Collect the corresponding eigenvalues
    lambdas = [eigvals[i] for i in range(n_components)]

    return alphas, lambdas
```

Ahora, creemos un nuevo conjunto de datos en forma de media luna y proyectémoslo en un subespacio unidimensional usando el kernel PCA de RBF:

```python
X, y = make_moons(n_samples=100, random_state=123)
alphas, lambdas = rbf_kernel_pca(X, gamma=15, n_components=1)
```

Para confirmar que nuestro enfoque produce los resultados correctos, supongamos que el punto 25 del conjunto de datos de la media luna es un nuevo punto de datos \\( \mathbf{x} \\), y queremos proyectarlo en este nuevo subespacio.

```python
x_new = X[25]
X_proj = alphas[25] # original projection
```

```python
x_new
```

```python
X_proj
```

```python
def project_x(x_new, X, gamma, alphas, lambdas):
    pair_dist = np.array([np.sum((x_new-row)**2) for row in X])
    k = np.exp(-gamma * pair_dist)
    return k.dot(alphas / lambdas)

# projection of the "new" datapoint
x_reproj = project_x(x_new, X, gamma=15, alphas=alphas, lambdas=lambdas) 
```
