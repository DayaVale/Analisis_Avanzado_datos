## Ejemplo de Suavizado mediante Kernel en 1D

```python
# Cargamos las librerias

import numpy as np

from numpy.linalg import inv

from numpy.random import uniform
from numpy.random import normal

import pandas as pd

from scipy.stats import norm

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt

%matplotlib inline
```

Creamos un conjunto de datos de la forma $Y=\sin (4 X)+\varepsilon, X \sim U[0,1], \varepsilon \sim N(0,1 / 3)$

```python
X = uniform(0, 1, 100)
X.sort()
Y = np.sin(4*X) + normal(0, 1/3, 100)
x_real = np.linspace(0, 1, 100)
y_real = np.sin(4*x_real)
```

```python
plt.plot(x_real, y_real, c='r', label='real')
plt.scatter(X, Y, s=2, label='data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend();
```

Construimos tres kernels: 
- KNN, 
- Epanechnikov: $D(t)=\left\{\begin{array}{cl}\frac{3}{4}\left(1-t^{2}\right) & \text { if }|t| \leq 1 \\ 0 & \text { otherwise }\end{array}\right.$
- Tricúbico:  $D(t)=\left\{\begin{array}{cl}\left(1-|t|^{3}\right)^{3} & \text { if }|t| \leq 1 \\ 0 & \text { otherwise }\end{array}\right.$

y ponderamos mediante Nadaraya-Watson:
$$K_{\lambda}\left(x_{0}, x\right)=D\left(\frac{\left|x-x_{0}\right|}{h_{\lambda}\left(x_{0}\right)}\right)$$
$$\hat{f}\left(x_{0}\right)=\frac{\sum_{i=1}^{N} K_{\lambda}\left(x_{0}, x_{i}\right) y_{i}}{\sum_{i=1}^{N} K_{\lambda}\left(x_{0}, x_{i}\right)}$$

```python
y_hat_30_nearest = np.zeros(y_real.shape)
y_hat_02_epa = np.zeros(y_real.shape)
y_hat_02_tri = np.zeros(y_real.shape)


for i in range(x_real.shape[0]):
    x0 = x_real[i]
    y_hat_30_nearest[i] = np.mean(Y[np.argsort(np.abs(X - x0))][:30])
    
    t = np.abs(X - x0)/0.2
    d = (t<=1)*3/4*(1-t**2) # kernel de Epanechnikov 
    d3 = (t<=1)*(1-t**3)**3 # kernel tricubico
    
    # Ponderacion mediante kernel Nadaraya-Watson
    y_hat_02_epa[i] = np.sum(d*Y)/np.sum(d)
    y_hat_02_tri[i] = np.sum(d3*Y)/np.sum(d3)
```

Y graficamos las suavizaciones:

```python
plt.figure(figsize=(10,6))
plt.plot(x_real, y_real, label='real')
plt.scatter(X, Y, s=2, label='data')
plt.plot(x_real, y_hat_30_nearest, label='KNN')
plt.plot(x_real, y_hat_02_epa, label='Epanechnikov')
plt.plot(x_real, y_hat_02_tri, label='Tri-cube')
plt.legend();
```

### Regresión local
Ahora implementaremos una regresión local usando un kernel de Epanechnikov. Recordemos que la regresión local se puede realizar mediante:
$$\hat{f}\left(x_{0}\right)=b\left(x_{0}\right)^{T}\left(\mathbf{B}^{T} \mathbf{W}\left(x_{0}\right) \mathbf{B}\right)^{-1} \mathbf{B}^{T} \mathbf{W}\left(x_{0}\right) \mathbf{y}$$

donde:

- $b(x)^{T}=(1, x)$ es la función con valores vectoriales.
- $B$ es la matriz de regresión $N \times 2$ con la fila i-ésima $b\left(x_{i}\right)^{T}$
- $\mathbf{W}\left(x_{0}\right)$ es la matriz diagonal $N \times N$ con el elemento diagonal $i$ $K_{\lambda}\left(x_{0}, x_{i}\right)$

```python
y_hat_lin = np.zeros(y_real.shape)
y_hat_sqr = np.zeros(y_real.shape)

B = np.hstack((np.ones(shape=(X.shape[0], 1)), np.expand_dims(X, axis=1)))
B2 = np.hstack((np.ones(shape=(X.shape[0], 1)), np.expand_dims(X, axis=1), np.expand_dims(X**2, axis=1)))

for i in range(x_real.shape[0]):
    x0 = x_real[i]
    t = np.abs(X - x0)/0.2
    d = (t<=1)*3/4*(1-t**2) # Epa kernel
    Wx0 = np.diag(d)
    y_hat_lin[i] = np.array([1, x0]) @ inv(B.T @ Wx0 @ B) @ B.T @ Wx0 @ np.expand_dims(Y, axis=1)
    y_hat_sqr[i] = np.array([1, x0, x0**2]) @ inv(B2.T @ Wx0 @ B2) @ B2.T @ Wx0 @ np.expand_dims(Y, axis=1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_real, y_real, label='real')
plt.scatter(X, Y, s=2, label='data')
plt.plot(x_real, y_hat_lin, label='Linear local Reg')
plt.plot(x_real, y_hat_sqr, label='Quadratic local Reg')
plt.legend();
```

## Estimación de la densidad mediante Kernel

Leemos el dataset de `Heart`, la información es la siguiente:

A retrospective sample of males in a heart-disease high-risk region of the Western Cape, South Africa. There are roughly two controls per case of CHD. Many of the CHD positive men have undergone blood pressure reduction treatment and other programs to reduce their risk factors after their CHD event. In some cases the measurements were made after these treatments. These data are taken from a larger dataset, described in  Rousseauw et al, 1983, South African Medical Journal. 

- sbp		systolic blood pressure
- tobacco		cumulative tobacco (kg)
- ldl		low densiity lipoprotein cholesterol
- adiposity
- famhist		family history of heart disease (Present, Absent)
- typea		type-A behavior
- obesity
- alcohol		current alcohol consumption
- age		age at onset
- chd		response, coronary heart disease

```python
df = pd.read_csv("heart.txt", index_col=0)
df.head()
```

Separamos $x$ y $y$

```python
x = df['sbp'].values
y = df['chd'].values
x_pos = x[y == 1]
x_neg = x[y == 0]
```

Determinamos el máximo y mínimo para $x$

```python
x_min, x_max = np.min(x), np.max(x)
```

Realizamos la estimación de la densidad usando `scipy.stats.norm.pdf()` y hallamos el promedio de la distribución centrada en cada punto. 

```python
x_real = np.linspace(x_min, x_max, 100)
y_real_pos = np.zeros(100)
y_real_neg = np.zeros(100)
for i in range(100):
    y_real_pos[i] = np.mean(norm.pdf(x_pos, x_real[i], 5.5)) # media x_real[i], desv_est 5.5 para la distribución
    y_real_neg[i] = np.mean(norm.pdf(x_neg, x_real[i], 5.5))
```

```python
plt.plot(x_real, y_real_neg, label='Chd negative')
plt.plot(x_real, y_real_pos, label='Chd positive')
plt.legend()
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Density Estimation');
```

**Estimación de las probabilidades posteriores para CHD mediante $$\frac{f_{k}(x) \pi_{k}}{\sum_{l=1}^{K} f_{l}(x) \pi_{l}}$$**

```python
# Estimacion de las probabilidades apriori usando proporciones
prior_pos = len(x_pos)/len(x) 
prior_neg = len(x_neg)/len(x)

y_pos_prob = np.zeros(100)

# Calculo de probabilidades posteriores usando Bayes
for i in range(100):
    y_pos_prob[i] = prior_pos*y_real_pos[i]/(prior_pos*y_real_pos[i] + prior_neg*y_real_neg[i])
    

plt.plot(x_real, y_pos_prob)
plt.ylim(0, 1)
plt.xlabel('Systolic Blood Pressure')
plt.ylabel('Density Estimation');
```

## Naive Bayes Classifier

El clasificador "Naive Bayes" asume que los descriptores (predictores) son independientes en cada clase. Es útil cuando el número de descriptores $p$ es grande.

El clasificador "Naive Bayes" se asemeja al clasificador discriminante lineal, pero asume que la matriz de covarianza $\Sigma_k$ es diagonal, en cuyo caso el discriminante resulta en:

$$
\delta(x)  \approx \log   \left[\pi_k  \prod_{j=1}^p f_{kj}(x_j)\right]=-\frac{1}{2} \sum_{j=1}^p \frac{( x_j - \mu_{kj} )^2}{\sigma^2} +\log \pi_k
$$

Este clasificador puede usar vectores *mixtos* (cualitativos y cuantitativos). Si $X_j$ es cualitativo, se remplaza $f_{kj}(x_j)$ con la función de probabilidad (para variables que toman valores discretos, es decir el histograma) sobre las categorías discretas.

A pesar de las grandes simplificaciones que hacer el clasificador *Naive Bayes* muchas veces produce buenos resultados.

```python
df2 = pd.get_dummies(df,drop_first=True)
df2.head()
```

```python
X = df2.drop(columns='chd')
y = df2['chd']
        
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
clf = GaussianNB()
clf.fit(X_train,y_train)
```

```python
print("La exactitud (accuracy) de clasificación con el clasificador de NB es:", clf.score(X_test, y_test))
```
