# Práctica: Clasificación

En esta primera parte se trabaja con el dataset `smarket.csv`, cuya descripción es la siguiente:

**The Stock Market Data**

Comenzaremos examinando algunos resúmenes numéricos y gráficos del conjunto de datos de **Smarket**, que forman parte de la biblioteca ISLR. Este *dataset* consta de rendimientos porcentuales para el índice de acciones [S&P 500](https://en.wikipedia.org/wiki/S%26P_500_Index) durante 1,250 días, desde principios de 2001 hasta finales de 2005. Para cada fecha, se han registrado los rendimientos porcentuales para cada uno de los cinco días de negociación anteriores, desde el 1 hasta el 5. También se ha registrado `Volume` (el número de acciones negociadas el día anterior, en miles de millones), `Today` (el porcentaje de rendimiento en la fecha en cuestión) y la variable que se quiere predecir `Direction` (si el mercado subió o bajó en esta fecha).

```python
# Importaremos la mayoria de modulos que se utilizaran a lo largo de la práctica

# Modulos basicos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

# Modulos de scikit-learn
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report, precision_score, accuracy_score
from sklearn.metrics import  roc_auc_score, roc_curve
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
# Funcion para calcular y visualizar la matriz de confusion 
def plotCM(ytrue, ypred, clases=None, normalize = False, ax = None):
    """ Funcion para calcular y visualizar la matriz de confusion"""
    
    if clases == None:
        clases = list(set(ytrue))
        clases.sort() # etiquetas unicas ordenadas alfabeticamente
    
    CM = confusion_matrix(ytrue,ypred, labels=clases)
    
    #Normaliza la matriz de confusion dividiendo cada fila por el total de verdaderos
    if normalize:
        CM = 100*CM / CM.sum(axis=1).reshape(-1,1) #Aprovechando el Broadcasting!
 
    df = pd.DataFrame(CM, index=clases, columns=clases)
    df.index.name = 'True'; df.columns.name = 'Predicted'
    
    sns.heatmap( df, # Visualizando la matriz de confusion
             annot=True, fmt='2.1f', cmap='ocean_r',cbar=False,square=True, annot_kws={'fontsize':16}, ax=ax )
    
    plt.show()
```

#### Importando el dataset de stock market 

```python
smarket = pd.read_csv('smarket.csv')
smarket.tail()
```

#### Organizando el dataset
Para realizar una predicción se tomará los años anteriores a 2005 como training y el 2005 para test. Es decir, se hará la predicción del año 2005 basado en los años anteriores.

```python
X = smarket.loc[:,'Lag1':'Volume']
y = smarket['Direction']

# Dividiendo en training y test
idx_train = smarket['Year'].values < 2005
X_train = X[idx_train]
X_test = X[~idx_train]
y_train = y[idx_train]
y_test = y[~idx_train]
```

## Regresión logística

```python
lrmodel = LogisticRegression(solver='lbfgs')
lrmodel.fit(X_train,y_train)
```

```python
# Matriz de confusion sobre el test
y_pred = lrmodel.predict(X_test)
plotCM(y_test, y_pred, normalize=True)
```

### Preprocesamiento

En muchas ocasiones es importante **preprocesar** los predictores. Por ejemplo, realizando una estandarización (restar la media y dividir por la desviación estándar). En scikit-learn se puede llamar mediante `sklearn.preprocessing.StandardScaler()`. En la documentación de [preprocessing](http://scikit-learn.org/stable/modules/preprocessing.h) se encuentran otros tipos de preprocesamiento y más información al respecto. A continuación se verá que sucede al escalar los predictores:

```python
# Se instancia al objeto StandarScaler para realizar el ajuste
# Esto es muy adecuado para preprocesar de la misma forma al test
scaler = preprocessing.StandardScaler().fit(X_train)

# Regresion logistica como antes pero escalando los predictores
lrmodel = LogisticRegression(solver='lbfgs')
lrmodel.fit( scaler.transform(X_train),y_train )

# Prediccion pero escalando los predictores
y_pred = lrmodel.predict( scaler.transform(X_test) )

# matriz de confusion sin normalizacion
plotCM(y_test,y_pred, normalize=True)
```

El error del test se puede calcular como $1 - accuracy$. Para el que desee saber más acerca de como imprimir valores en python puede ir a [este excelente tutorial](https://pyformat.info/).

```python
# Test error
print("El accuracy del test es: {:.2f}".format( accuracy_score(y_test,y_pred) ))
test_error = 1-accuracy_score(y_test,y_pred)
print("El error del test es: {:.2f}".format( test_error ) )
```

### Feature selection
Con `statmodels` hemos hecho algo sobre selección de predictores, o llamado en inglés, *feature selection*. En ese caso, hemos visto que el *p-value* al realizar una clasificación de regresión logística, era inverso a la relevancia del predictor. `scikit-learn` tiene el módulo [*sklearn.feature_selection*](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.feature_selection), el cuál tiene diversos métodos implementados para selección de descriptores. 

A continuación usaremos uno de los más simples, `sklearn.feature_selection.f_classif(X, y)` que realiza un análisis ANOVA con la estadística F (retornando también p-values) de cada predictor, respecto a las clases (es el método más similar al que hemos usado con statmodels). 

Una explicación rápida e intuitiva de la estadística F podéis encontrarla en este [link](https://www.statisticshowto.datasciencecentral.com/probability-and-statistics/f-statistic-value-test/)

```python
from sklearn import feature_selection

F,p = feature_selection.f_classif(X_train, y_train)

# Crea un dataframe con los F-value y p-value para visualizar mejor
pd.DataFrame(np.array([F,p]).T,index=X_train.columns, columns=['F_value','p_value'])
```

¿Qué sucede si se realiza el ajuste considerando sólo los predictores con valor-p más alto, es decir, `Lag1` y `Lag2`?

```python
# Tomando de nuevo los datos
X = smarket[['Lag1','Lag2']]
y = smarket['Direction'] # No es necesario para sklearn codificar la variable de salida

# Dividiendo en training y test
idx_train = smarket['Year'].values < 2005
X_train = X[idx_train]
X_test = X[~idx_train]
y_train = y[idx_train]
y_test = y[~idx_train]

# Creacion del modelo y prediccion con preprocesamiento
scaler = preprocessing.StandardScaler().fit(X_train)
lrmodel = LogisticRegression(solver='lbfgs')
lrmodel.fit( scaler.transform(X_train),y_train )
y_pred = lrmodel.predict( scaler.transform(X_test) )

# matriz de confusion 
plotCM(y_test,y_pred, normalize=True)
```

Se nota una mejora, que se puede verificar con el accuracy:

```python
print("El accuracy del test es: {:.2f}".format( accuracy_score(y_test,y_pred) ))
```

## Análisis discriminante lineal

Se realiza el análisis con el anterior $X$ e $y$, es decir, considerando solamente los dos días previos. 

```python
# Se instancia el objeto del modelo LDA
lda_model = LinearDiscriminantAnalysis()

# Ajuste con estandarizacion de X_train
lda_model.fit( scaler.transform(X_train), y_train )

# Prediccion con la estandarizacion de X_test
y_pred = lda_model.predict( scaler.transform(X_test) )

# matriz de confusion sin normalizacion
plotCM(y_test,y_pred,normalize=True)
```

*¿Por qué el resultado es el mismo que en regresión logística?*

Las probabilidades anteriores (prior probabilities) son:

```python
lda_model.priors_
```

Las medias para cada clase por descriptor son:

```python
lda_model.means_(en el libro ISLR)
```

Los coeficientes del modelo LDA (con los datos preprocesados) son:

```python
lda_model.coef_
```

Las probabilidades a priori son π1 = 0.492 y π2 = 0.508; en otras palabras, el 49.2% de las observaciones de entrenamiento corresponden a días durante los cuales el mercado bajó. También proporciona las medias del grupo; es decir, el promedio de cada predictor dentro de cada clase, y el modelo de LDA los utiliza como estimaciones de μk. Esto sugiere que hay una tendencia a que los retornos de los 2 días anteriores sean negativos en los días en que el mercado aumenta, y una tendencia a que los retornos de los días anteriores sean positivos en los días en que el mercado declina. Los coeficientes de los discriminantes lineales proporcionan la combinación lineal de Lag1 y Lag2 que se utilizan para formar la regla de decisión del modelo LDA.
En otras palabras, estos multiplican a los predictores en X = x. Si -0.068×Lag1 - 0.055×Lag2 es grande, entonces el clasificador LDA predecirá un incremento en el mercado, y si es pequeño, entonces el clasificador LDA predecirá un declive en el mercado.

## Análisis discriminante cuadrático

```python
qda_model = QuadraticDiscriminantAnalysis() #Instancia al modelo
qda_model.fit( scaler.transform(X_train), y_train ) # Ajuste o entrena
y_pred = qda_model.predict( scaler.transform(X_test) ) # prediccion
plotCM(y_test, y_pred,normalize=True) # Matriz de confusion
```

## Naive Bayes

```python
nb_model = GaussianNB()
nb_model.fit(X_train,y_train)
y_pred = nb_model.predict(X_test) # Prediccion
plotCM(y_test,y_pred) # Matriz de confusion
```

El conjunto de datos contiene 50 muestras de cada una de tres especies de Iris (Iris setosa, Iris virginica e Iris versicolor). Carguemos el dataset de Iris desde seaborn (se puede descargar de muchos lugares):

Definiendo $X$ e $y$

```python
iris = sns.load_dataset("iris")
iris.sample(10)
```

```python
sns.pairplot(iris, hue="species");
```

Dividiendo los datos aleatoriamente en training y test:

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,0:4],iris['species'],
                                                   test_size=0.25, random_state=123)
```

#### Regresión logística

```python
model = LogisticRegression(solver='lbfgs', multi_class='ovr').fit(X_train,y_train)
y_pred = model.predict(X_test)
plotCM(y_test,y_pred, normalize=True)
```

Reporte de clasificación. Recordad que *recall* es *sensibilidad*. Definiciones más concretas de cada término la podéis encontrar en este [link](https://en.wikipedia.org/wiki/Confusion_matrix).

```python
print(classification_report(y_test,y_pred))
```

### LDA

```python
model = LinearDiscriminantAnalysis().fit( X_train, y_train )
y_pred = model.predict(X_test)
plotCM(y_test,y_pred, normalize=True)
```

### QDA

```python
model = QuadraticDiscriminantAnalysis().fit( X_train, y_train )
y_pred = model.predict(X_test)
plotCM(y_test,y_pred, normalize=True)
```

### KNN

```python
model = KNeighborsClassifier(n_neighbors=3).fit( X_train, y_train )
y_pred = model.predict(X_test)
plotCM(y_test,y_pred, normalize=True)
```

### Naive Bayes

```python
model = GaussianNB().fit( X_train, y_train )
y_pred = model.predict(X_test)
plotCM(y_test,y_pred, normalize=True)
```

Se concluye que el mejor clasificador puede ser la regresión logística, o LDA o QDA. Ya que presentan resultados similares.
En este ejemplo, los descriptores son longitudes con las mismas dimensiones, por lo tanto no hay necesidad de realizar preprocesamiento (específicamente escalamiento). Sin embargo, cuando los descriptores tienen unidades diferentes u órdenes de longitud muy dispares es necesario realizar algún tipo de escalamiento (en inglés feature scaling).
